```python
import pickle
import pandas as pd

with open("results/ablation_imdb_distilbert_kan-1.pickle", "rb") as f:
    imdb_distilbert_kan_1 = pickle.load(f)

with open("results/ablation_imdb_distilbert_kan-3.pickle", "rb") as f:
    imdb_distilbert_kan_3 = pickle.load(f)

with open("results/imdb_distilbert_kan.pickle", "rb") as f:
    imdb_distilbert_kan_5 = pickle.load(f)

with open("results/imdb_distilbert_kan--7.pickle", "rb") as f:
    imdb_distilbert_kan_7 = pickle.load(f)

with open("results/imdb_distilbert_kan--9.pickle", "rb") as f:
    imdb_distilbert_kan_9 = pickle.load(f)


imdb_distilbert_kan_epoch = pd.DataFrame(
    [
        [
            1,
            imdb_distilbert_kan_1["test_metrics"].overall_stat["Overall ACC"],
            imdb_distilbert_kan_1["test_metrics"].overall_stat["F1 Macro"],
        ],
        [
            3,
            imdb_distilbert_kan_3["test_metrics"].overall_stat["Overall ACC"],
            imdb_distilbert_kan_3["test_metrics"].overall_stat["F1 Macro"],
        ],
        [
            5,
            imdb_distilbert_kan_5["test_metrics"].overall_stat["Overall ACC"],
            imdb_distilbert_kan_5["test_metrics"].overall_stat["F1 Macro"],
        ],
        [
            7,
            imdb_distilbert_kan_7["test_metrics"].overall_stat["Overall ACC"],
            imdb_distilbert_kan_7["test_metrics"].overall_stat["F1 Macro"],
        ],
        [
            9,
            imdb_distilbert_kan_9["test_metrics"].overall_stat["Overall ACC"],
            imdb_distilbert_kan_9["test_metrics"].overall_stat["F1 Macro"],
        ],
    ],
    columns=["grid_size", "Overall ACC", "F1 Macro"],
)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(5, 3.5))

sns.lineplot(
    x="grid_size",
    y="Overall ACC",
    data=imdb_distilbert_kan_epoch,
    color="steelblue",
    label="Overall ACC",
    marker="o",
    linewidth=1.5
)
sns.lineplot(
    x="grid_size",
    y="F1 Macro",
    data=imdb_distilbert_kan_epoch,
    color="coral",
    label="F1 Macro",
    marker="o",
    linewidth=1.5
)

# plt.title("Distilbert + KAN on IMDB by Grid Size")
plt.xlabel("Grid Size")
plt.ylabel("Score")

plt.yticks([0.50, 0.55, 0.60, 0.65, 0.70, 0.75, 0.80, 0.85, 0.90, 0.95, 1.00])
plt.xticks([1, 2, 3, 4, 5, 6, 7, 8, 9])

plt.grid(color="slategray", linestyle="dashed", linewidth=0.5, zorder=0)
plt.legend(loc="best")
plt.tight_layout()
plt.savefig("imdb-distilbert-kan-acc-f1-by-grid-size.pdf", format="pdf", dpi=300)
plt.savefig("imdb-distilbert-kan-acc-f1-by-grid-size.eps", format="eps", dpi=300)

plt.show();
```

```python
import pandas as pd

per_epoch_data = pd.read_csv("per-epoch-stats.csv")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

def plot_mlp_kan_stat_per_epoch(data, dataset, best_model, stat_name):
    plt.figure(figsize=(5, 3.5))

    sns.lineplot(
        x="Epoch",
        y=stat_name,
        data=data.query(
            f"dataset=='{dataset}' and  best_model=='{best_model}' and classifier=='kan'"
        ),
        color="steelblue",
        label="KAN",
        marker="o",
        linewidth=2.0,
    )
    sns.lineplot(
        x="Epoch",
        y=stat_name,
        data=data.query(
            f"dataset=='{dataset}' and  best_model=='{best_model}' and classifier=='mlp'"
        ),
        color="coral",
        label="MLP",
        marker="o",
        linewidth=2.0,
    )

    plt.xlabel("Epoch")
    plt.ylabel(stat_name)

    # plt.yticks([0.50, 0.60, 0.70, 0.80, 0.90, 1.00])
    plt.xticks([1, 2, 3, 4, 5])

    plt.grid(color="slategray", linestyle="dashed", linewidth=0.5, zorder=0)
    plt.legend(loc="best")

    plt.tight_layout()
    plt.savefig(
        f"plots/{dataset}-{best_model}-{stat_name.lower().replace(' ','-')}-per-epoch.pdf",
        format="pdf",
        dpi=300,
    )
    plt.savefig(
        f"plots/{dataset}-{best_model}-{stat_name.lower().replace(' ','-')}-per-epoch.eps",
        format="eps",
        dpi=300,
    )

    plt.show();
```
